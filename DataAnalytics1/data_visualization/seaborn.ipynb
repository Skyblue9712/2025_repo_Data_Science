```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.shape
```

```python
sns.histplot(df, x='age', bins=10, color='red', kde=True)
```

```python
import matplotlib.pyplot as plt
sns.histplot(df, x='age', bins=10, color='red', kde=True)
plt.xticks(range(0, 81, 5))
plt.show()
```

```python
sns.countplot(df, x='sex', palette="rocket")
plt.title('Number of Passangers by Gender')
plt.show()
```

```python
sns.countplot(df, x="pclass", hue="sex")
plt.title('Number of Passangers by Class')
plt.show()
```

```python
sns.scatterplot(df, x='age', y='fare', hue='sex', palette='magma')
plt.title('Age vs Fare by Gender')
plt.show()
```

```python
sns.scatterplot(df, x='age', y='fare', hue='alive', size= 'pclass', palette='magma')
plt.title('Age vs Fare by Gender')
plt.show()
```
