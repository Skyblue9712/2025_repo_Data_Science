```python
from sqlalchemy import create_engine,Column, Integer, String
from sqlalchemy.orm import declarative_base, sessionmaker
```

```python
engine = create_engine('sqlite:///database.db')
```

```python
Base = declarative_base()
```

```python
class User(Base):
    __tablename__ = 'users'
    id = Column(Integer, primary_key=True)
    name = Column(String)
    email = Column(String)
    password= Column(String)
```

```python
Base.metadata.create_all(engine)
```

```python
Session= sessionmaker(bind=engine)
```

```python
session = Session()
```

```python
new_user = User(name='Hood Hogan',email='hood@mail.com', password='hogan123')
session.add(new_user)
session.commit()
```

```python
new_user = User(name='Sky Blue',email='Sky@mail.com', password='Sky@123')
session.add(new_user)
session.commit()
```

```python
users = session.query(User).all() # Fetch all users
users
```

```python
for u in users:
    print(f'ID: {u.id}, Name: {u.name}, Password: {u.password}')
```

```python
user = session.query(User).filter_by(email='Sky@mail.com').first()
print(user.name, user.password)
```

```python
user.name= 'Farhan'
session.commit()
```

```python
class Student(Base):
    __tablename__ = 'Students'
    roll_no = Column(Integer, primary_key=True)
    name = Column(String)
    course = Column(String)
    email= Column(String)
```

```python
Base.metadata.create_all(engine)
```

```python
Session= sessionmaker(bind=engine)
```

```python
session = Session()
```

```python
new_student = Student(name='Hood Hogan',course='B.Sc.',email='hood@mail.com')
session.add(new_student)
session.commit()
```

```python
new_student = Student(name='Sky Blue',course='B.Com.',email='Sky@mail.com')
session.add(new_student)
session.commit()
```

```python
new_student = Student(name='Shivam Tripathi',course='B.Sc.',email='Tripathi@mail.com')
session.add(new_student)
session.commit()
```

```python
user= session.query(Student).filter_by(course='B.Sc.').all()
for u in user:
    print(f'Roll No: {u.roll_no}, Name: {u.name}, Course: {u.course}, Email: {u.email}')
```
