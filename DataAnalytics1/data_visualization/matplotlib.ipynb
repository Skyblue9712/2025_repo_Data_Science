```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('immigration_data.csv')
df.head()
```

```python
df.set_index("Country", inplace=True)
df.head()
```

```python
plt.figure(figsize=(15,5))
years=[str(year) for year in range(1980, 2014)]
india = df[years].loc['India']
plt.plot(years,india)
plt.xticks(years[::2], rotation=45)
```

```python
years = map(str, range(1980, 2014))
india = df[years].loc['India']
india.head()
```

```python
plt.plot(india.index, india.values)
plt.show()
```

```python
india.plot(figsize=(15, 5), rot=45)
plt.xlabel('Year')
plt.ylabel('No. of Immigrants')
plt.title('Indian Immigrants Data')
plt.show()
```

```python
top_10 = df.sort_values('2013', ascending=False)['2013'].head(10)
top_10
```

```python
top_10.plot(kind='bar', figsize=(20, 5), rot=60)
plt.tight_layout()
plt.show()
```
