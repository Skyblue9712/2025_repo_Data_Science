```python
strNums = ['45', '8765', '123', '9']
output =  map( lambda s : int(s) , strNums )
```

```python
list(output)
```

```python
cleaned_rupees = map( lambda s : int(s[1:].split('.')[0]) , rupees )
list(cleaned_rupees)
```

```python
prices = [2000, 4500, 1200, 400, 6000, 7500]
```

```python
filtered_prices =  filter( lambda p : p > 1000 and p < 5000, prices )
list(filtered_prices)
```
