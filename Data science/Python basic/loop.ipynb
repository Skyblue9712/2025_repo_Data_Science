```python
num = int(input('enter a number: '))

rev = 0

while num>0:
    d = num%10
    rev = rev * 10 + d
    num = num//10
    
print('reversed number is:', rev)
```

```python
# WAP to check if number is palindrome or not

num = int(input('enter a number: '))
temp = num
rev = 0

while num>0:
    d = num%10
    rev = rev * 10 + d
    num = num//10
   
print(num)
if rev == temp:
    print('palindrome')
else:
    print('not palindrome')
```
