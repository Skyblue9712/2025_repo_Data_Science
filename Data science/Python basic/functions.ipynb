```python
# variable length parameter functions

def addAll(*nums):
    res = sum(nums)
    
    print(res)
```

```python
addAll(38, 78, 93, 19, 21)
```

```python
def getPercentage(*marks, total = 100):
    obt = sum(marks)
    percent = obt * 100 /(len(marks) * total)
    
    print(percent)
```

```python
getPercentage(50, 67, 87)
```

```python
def genNums(n):
    for i in range(1, n+1):
        yield i
```

```python
for i in genNums(20):
    print(i)
```

```python
gen20 = genNums(20)
```

```python
for i in gen20:
    print(i)
```
