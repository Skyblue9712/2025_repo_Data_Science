```python
class Dog:
    
    def __init__(self, name, breed, nature, toy):
        self.name = name
        self.breed = breed
        self.nature = nature
        self.toy = toy
        
    def play(self):
        print(f'{self.name} plays with {self.toy}')
        
    def bark(self):
        print(f'{self.name} barks loudly')
```

```python
dogesh = Dog('Lord Dogesh', 'German Shepherd', 'Calm', 'Ball')
```

```python
dogesh.nature
```

```python
dogesh.name
```

```python
dogesh.play()
```

```python
tommy = Dog('Tommy', 'Golden Retriver', 'Aggressive', 'Frisbee')
```

```python
tommy.bark()
```
