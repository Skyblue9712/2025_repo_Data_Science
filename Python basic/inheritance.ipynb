```python
class Laptop:
    def __init__(self, brand, model, cpu, price):
        self.brand = brand
        self.model = model
        self.cpu = cpu
        self.price = price
        
    def turnOn(self):
        return f"{self.brand} {self.model} is now ON"
    
    def playGame(self, game_name):
        return f"Playing {game_name} on {self.brand} {self.model}"
```

```python
class GamingLaptop(Laptop):
    def __init__(self, brand, model, cpu, price, gpu, keyboard):
        super().__init__(brand, model, cpu, price)
        self.gpu = gpu
        self.keyboard = keyboard
        
    def turnOn(self):
        return "Gaming laptop turns on quickly"
        
```

```python
lappy = GamingLaptop('Lenevo', 'LOQ', 'Ryzen 7', 83000, 'RTX 4060', "white backlit")
```

```python
lappy.turnOn()
```

```python
lappy2 = Laptop('Asus', 'Vivobook 14', 'Ryzen 5', 36000)
```
