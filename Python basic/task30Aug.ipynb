```python
# WAP to check if number is palindrome or not

num = int(input('enter a number: '))
temp = num
rev = 0

while num>0:
    d = num%10
    rev = rev  + d ** 3
    num = num//10
   
print(rev)

if rev == temp:
    print('armstrong number')
else:
    print('not a armstrong number')
```
