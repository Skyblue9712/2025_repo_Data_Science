```python
# series can also be created from list,tupples,and dictionaries
```

```python
import pandas as pd
import numpy as np

list3= ['a','e','i','o','u']
tup3 =('akash','deepak','rohit')
dic3 = {'a':1,'b':2,'c':3}

ser3 = pd.Series(list3)
ser4 = pd.Series(tup3)
ser5 = pd.Series(dic3)
print(ser3)
print(ser4)

print(ser5)
```
